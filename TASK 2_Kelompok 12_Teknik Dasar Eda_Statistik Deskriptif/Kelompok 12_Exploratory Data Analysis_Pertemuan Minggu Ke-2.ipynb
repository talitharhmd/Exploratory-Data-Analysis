##### Bagaimana ukuran penyebaran dan pemusatan data terkait capaian pendapatan target dan jumlah pajak ? 

```python
print('-----------Capaian Pendapatan------------')

# Menghitung mean
mean = df['capaian pendapatan'].mean()
# Menghitung median
median = df['capaian pendapatan'].median()
# Menghitung modus
mode = df['capaian pendapatan'].mode().iloc[0]
range_umur = df['capaian pendapatan'].max() - df['capaian pendapatan'].min()
varians_umur = df['capaian pendapatan'].var()
std_dev_umur = df['capaian pendapatan'].std()

print("Range capaian pendapatan:", range_umur)
print("Varians capaian pendapatan:",varians_umur)
print("Std capaian pendapatan:", std_dev_umur)
print("Mean capaian pendapatan:", mean)
print("Median capaian pendapatan:", median)
print("Modus capaian pendapatan:", mode)

print('-----------Jumlah------------')

# Menghitung mean
mean_jml = df['jumlah'].mean()
# Menghitung median
median_jml = df['jumlah'].median()
# Menghitung modus
mode_jml = df['jumlah'].mode().iloc[0]
range_jml = df['jumlah'].max() - df['jumlah'].min()
varians_jml = df['jumlah'].var()
std_dev_jml = df['jumlah'].std()

print("Range Jumlah:", range_jml)
print("Varians Jumlah:",varians_jml)
print("Std Jumlah:", std_dev_jml)
print("Mean Jumlah:", mean_jml)
print("Median Jumlah:", median_jml)
print("Modus Jumlah:", mode_jml)
```

Data capaian pendapatan cukup bervariasi, ditunjukkan oleh nilai varians yang cukup tinggi.
Nilai median yang dekat dengan nilai rata-rata menunjukkan bahwa distribusi capaian pendapatan cenderung simetris, meskipun ada beberapa nilai ekstrem yang menggeser rata-rata ke atas.Modus yang lebih rendah dari rata-rata dan median menunjukkan bahwa ada beberapa nilai yang lebih rendah yang muncul lebih sering dalam data ini.

Data jumlah memiliki variasi yang sangat tinggi, ditunjukkan oleh nilai varians dan standar deviasi yang sangat besar.
Median yang jauh lebih rendah dari rata-rata menunjukkan adanya nilai ekstrem yang sangat tinggi dalam data jumlah.
Modus yang sangat rendah menunjukkan bahwa ada nilai-nilai yang sangat sering muncul dalam data ini, sementara sebagian besar nilai sangat tinggi. Hal ini mengindikasikan ketidakseimbangan yang signifikan dalam distribusi data jumlah.


## 4. Apakah terdapat perbedaan yang signifikan dalam penerimaan pajak antara hari kerja dan hari libur?

```python
import matplotlib.pyplot as plt

total_penerimaan_hari_kerja = df[df['prosen'] > 0]['jumlah'].sum()
total_penerimaan_hari_libur = df[df['prosen'] == 0]['jumlah'].sum()

plt.figure(figsize=(8,6))
plt.bar(['Hari Kerja', 'Hari Libur'], [total_penerimaan_hari_kerja, total_penerimaan_hari_libur], color=['#C08261', '#E2C799'])
plt.title('Perbandingan Penerimaan Pajak pada Hari Kerja dan Hari Libur')
plt.xlabel('Hari')
plt.ylabel('Total Pajak')
plt.show()
```

Terdapat perbedaan yang signifikan dalam penerimaan pajak antara hari kerja dan hari libur. Penerimaan pajak pada hari kerja jauh lebih tinggi daripada pada hari libur. Hasil ini menunjukkan bahwa penerimaan pajak cenderung terjadi lebih sering dilakukan pada hari kerja daripada hari libur.

## 5. Apakah terdapat perbedaan dalam Capaian Pendapatan wajib pajak per Wilayah pada 3 bulan terakhir pada Tahun 2022 di Jawa Timur? 

```python
df1 = df.copy()
```

```python
wilayah = {
    'SURABAYA BARAT': 'SURABAYA',
    'SURABAYA TIMUR': 'SURABAYA',
    'SURABAYA UTARA': 'SURABAYA',
    'SURABAYA SELATAN': 'SURABAYA', 
    'GRESIK': ' GRESIK', 
    'SIDOARJO': 'SIDOARJO', 
    'MOJOKERTO': 'MOJOKERTO', 
    'JOMBANG': 'JOMBANG',
    'BOJONEGORO': 'BOJONEGORO', 
    'LAMONGAN': 'LAMONGAN', 
    'TUBAN': 'TUBAN', 
    'MADIUN': 'MADIUN', 
    'NGAWI': 'NGAWI', 
    'MAGETAN': 'MAGETAN',
    'PONOROGO': 'PONOROGO', 
    'PACITAN': 'PACITAN', 
    'KEDIRI': 'KEDIRI',
    'BLITAR': 'BLITAR', 
    'TULUNGAGUNG': 'TULUNGAGUNG',
    'TRENGGALEK': 'TRENGGALEK', 
    'NGANJUK': 'NGANJUK', 
    'MALANG KOTA': 'MALANG', 
    'MALANG UTARA': 'MALANG',
    'MALANG SELATAN': 'MALANG', 
    'PASURUAN': 'PASURAN', 
    'PROBOLINGGO': 'PROBOLINGGO', 
    'LUMAJANG': 'LUMAJANG', 
    'JEMBER': 'JEMBER',
    'BONDOWOSO': 'BONDOWOSO',
    'SITUBONDO': 'SITUBONDO',
    'BANYUWANGI': 'BANYUWANGI', 
    'PAMEKASAN': 'MADURA', 
    'SAMPANG': 'MADURA',
    'BANGKALAN': 'MADURA', 
    'SUMENEP': 'MADURA'
}

# Mengganti nilai nama bulan dengan angka menggunakan fungsi map
df1['Wilayah'] = df1['upt'].map(wilayah)
df1.drop(columns= ['upt'], inplace=True)
df1
```

```python
df2 = df1.copy()
```

```python
data = df2.groupby('Wilayah')['capaian pendapatan'].sum()

capaian_pendapatan = pd.DataFrame(data)
capaian_pendapatan = capaian_pendapatan.sort_values(by='capaian pendapatan', ascending=False)
capaian_pendapatan
```

```python
import matplotlib.pyplot as plt

# Membuat bar chart
plt.figure(figsize=(15, 8))
capaian_pendapatan.plot(kind='bar', color='#7C81AD')
plt.xlabel('Wilayah')
plt.ylabel('Total Penerimaan Pajak')
plt.title('Capaian Pendapatan Pajak di Jawa Timur')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
capaian_pendapatan_sorted = capaian_pendapatan.sort_values(by='capaian pendapatan', ascending=False)

tinggi_5 = capaian_pendapatan_sorted
rendah_5 = capaian_pendapatan_sorted.tail()

colors_tinggi = sns.color_palette("GnBu_d", len(tinggi_5))
colors_rendah = sns.color_palette("YlOrBr_d", len(rendah_5))

# Mengambil 5 wilayah dengan tingkat kepatuhan tertinggi
tinggi_5 = capaian_pendapatan_sorted.head()
tinggi_5

# Membuat bar chart untuk 5 wilayah dengan tingkat kepatuhan tertinggi
plt.figure(figsize=(10, 6))
tinggi_5.plot(kind='barh', color=colors_tinggi)
plt.xlabel('Total Penerimaan Pajak')
plt.ylabel('Wilayah')
plt.title('5 Wilayah dengan Capaian Pendapatan Tertinggi Wajib Pajak di Jawa Timur')
plt.tight_layout()
plt.show()

rendah_5 = capaian_pendapatan_sorted.tail()
rendah_5

plt.figure(figsize=(10, 6))
rendah_5.plot(kind='barh', color=colors_rendah)
plt.xlabel('Wilayah')
plt.ylabel('Total Penerimaan Pajak')
plt.title('5 Wilayah dengan Capaian Pendapatan Terendah Wajib Pajak di Jawa Timur')

plt.tight_layout()
plt.show()
```
